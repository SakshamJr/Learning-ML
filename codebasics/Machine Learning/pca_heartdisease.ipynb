```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Datasets\heart.csv')
data.head()
```

```python
data.columns
```

```python
to_encode = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]
```

```python
dummies = pd.get_dummies(data[to_encode])
```

```python
dummies.columns
```

```python
dummies
```

```python
dummies.drop(
    [
        "Sex_F",
        "ChestPainType_TA",
        "RestingECG_ST",
        "ExerciseAngina_N",
        "ST_Slope_Down",
    ],
    axis=1,
    inplace=True,
)
```

```python
dummies
```

```python
data
```

```python
data.drop(
    [
        "Sex",
        "ChestPainType",
        "RestingECG",
        "ExerciseAngina",
        "ST_Slope",
        "FastingBS"
    ], axis=1, inplace=True
)
```

```python
data
```

```python
X = data.drop("HeartDisease", axis=1)
y = data.HeartDisease
X = pd.concat([X, dummies], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=10
)
```

```python
X_train
```

```python
X_train.shape
```

```python
num = ["Age", "RestingBP", "Cholesterol", "MaxHR", "Oldpeak"]
cat = [
    "Sex_M",
    "ChestPainType_ASY",
    "ChestPainType_ATA",
    "ChestPainType_NAP",
    "RestingECG_LVH",
    "RestingECG_Normal",
    "ExerciseAngina_Y",
    "ST_Slope_Flat",
    "ST_Slope_Up",
]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Scale numerical features and reset indices
X_train_scaled_num = pd.DataFrame(scaler.fit_transform(X_train[num]), columns=X_train[num].columns).reset_index(drop=True)
X_train_cat = X_train[cat].reset_index(drop=True)

# Concatenate the scaled numerical features with the categorical features
X_train_scaled = pd.concat([X_train_scaled_num, X_train_cat], axis=1)
print(X_train_scaled.shape)

# Similarly, for the test set
X_test_scaled_num = pd.DataFrame(scaler.transform(X_test[num]), columns=X_test[num].columns).reset_index(drop=True)
X_test_cat = X_test[cat].reset_index(drop=True)

X_test_scaled = pd.concat([X_test_scaled_num, X_test_cat], axis=1)
print(X_test_scaled.shape)
```

```python
X_train_scaled.shape
```

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

model_params = {
    "svm": {
        "model": SVC(gamma="auto"),
        "params": {"C": [1, 5, 10], "kernel": ["rbf", "linear", "poly"]},
    },
    "logistic": {
        "model": LogisticRegression(
            solver="liblinear", multi_class="auto", max_iter=500
        ),
        "params": {"C": [1, 5, 10]},
    },
    "xgboost": {
        "model": XGBClassifier(),
        "params": {
            "n_estimators": [25, 50, 75, 100],
            "learning_rate": [0.01, 0.05, 0.1],
        },
    },
    "randomforest": {
        "model": RandomForestClassifier(),
        "params": {"n_estimators": [10, 30, 50]},
    },
}

scores = []
for model_name, mp in model_params.items():
    model = GridSearchCV(mp["model"], mp["params"], cv=5, return_train_score=False)
    model.fit(X_train_scaled, y_train)
    scores.append(
        {
            "model": model_name,
            "best_score": model.best_score_,
            "best_params": model.best_estimator_,
        }
    )
```

```python
df = pd.DataFrame(scores, columns = ['model','best_score', 'best_params'])
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=7)
```

```python
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
X_train_pca_df = pd.DataFrame(X_train_pca, columns=[f'PC{i+1}' for i in range(X_train_pca.shape[1])])
X_test_pca_df = pd.DataFrame(X_test_pca, columns=[f'PC{i+1}' for i in range(X_test_pca.shape[1])])

# 4. Reset indices of categorical features
X_train_cat_reset = X_train[cat].reset_index(drop=True)
X_test_cat_reset = X_test[cat].reset_index(drop=True)

# 5. Concatenate PCA components with the original categorical features
X_train_scaled = pd.concat([X_train_pca_df, X_train_cat_reset], axis=1)
X_test_scaled = pd.concat([X_test_pca_df, X_test_cat_reset], axis=1)
```
