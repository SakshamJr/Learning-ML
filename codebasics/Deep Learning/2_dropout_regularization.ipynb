```python
X = df.drop(60, axis = 1)
y = df[60]
```

```python
y = pd.get_dummies(y)
y.drop('M',axis=1 ,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=10, test_size=0.2
)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential(
    [
        keras.layers.Input(shape=(60,)),
        keras.layers.Dense(60, activation="relu"),
        keras.layers.Dense(30, activation="relu"),
        keras.layers.Dense(15, activation="relu"),
        keras.layers.Dense(1, activation="sigmoid"),
    ]
)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
y_pred = np.round(y_pred)
print(classification_report(y_pred, y_test))
```

```python
modeld = keras.Sequential(
    [
        keras.layers.Input(shape=(60,)),
        keras.layers.Dense(60, activation="relu"),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(30, activation="relu"),
        keras.layers.Dropout(0.4),
        keras.layers.Dense(15, activation="relu"),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(1, activation="sigmoid"),
    ]
)

modeld.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
from sklearn.metrics import classification_report

y_predd = modeld.predict(X_test)
y_predd = np.round(y_predd)
print(classification_report(y_predd, y_test))
```
