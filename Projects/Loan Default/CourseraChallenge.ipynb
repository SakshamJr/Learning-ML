```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_df = pd.read_csv('train.csv')
```

```python
train_df.head()
```

```python
train_df['LoanPurpose'].value_counts()
```

```python
columns_to_encode = ['Education', 'EmploymentType', 'MaritalStatus', 'HasMortgage', 'HasDependents', 'LoanPurpose', 'HasCoSigner']
# Perform One-Hot Encoding
train_df_encoded = pd.get_dummies(train_df, columns=columns_to_encode)

# Display the first few rows of the new DataFrame
train_df_encoded.head()
```

```python
train_df_encoded.drop('LoanID',axis=1).corr()['Default'].sort_values(ascending=False)
```

```python
train_df_clean = train_df_encoded.drop(['LoanID','Education_PhD','EmploymentType_Full-time',
                                 'MaritalStatus_Single','HasMortgage_Yes','HasDependents_Yes',
                                 'LoanPurpose_Auto','HasCoSigner_Yes'],axis=1)
# y_train = train_df_encoded['Default']
```

```python
train_df_clean.info()
```

```python
train_df_clean.corr()['Default'].sort_values(ascending=False)
```

```python
train_df_clean['SimpleInterest'] = train_df_clean['LoanAmount'] * train_df_clean['InterestRate'] * train_df_clean['LoanTerm'] / 100
train_df_clean.corr()['Default'].sort_values(ascending=False)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = train_df_clean.drop('Default',axis=1)
y_train = train_df_clean['Default']
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_train = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
test_df = pd.read_csv('test.csv')
```

```python
test_df.head()
```

```python
columns_to_encode = ['Education', 'EmploymentType', 'MaritalStatus', 'HasMortgage', 'HasDependents', 'LoanPurpose', 'HasCoSigner']
# Perform One-Hot Encoding
test_df_encoded = pd.get_dummies(test_df, columns=columns_to_encode)
test_df_clean = test_df_encoded.drop(['LoanID','Education_PhD','EmploymentType_Full-time',
                                 'MaritalStatus_Single','HasMortgage_Yes','HasDependents_Yes',
                                 'LoanPurpose_Auto','HasCoSigner_Yes'],axis=1)
```

```python
test_df_clean['SimpleInterest'] = test_df_clean['LoanAmount'] * test_df_clean['InterestRate'] * test_df_clean['LoanTerm'] / 100
```

```python
y_pred = model.predict(test_df_clean)
for i in range (50):
    print(y_pred[i],end=' ')
```
